```python
import numpy as np
import pandas as pd
import pathlib
import os
from matplotlib import pyplot as plt, image
from os import walk
from PIL import Image as im
from collections import Counter 
from itertools import chain 
# from IPython.display import Image
```

```python
# path
home = pathlib.Path.cwd().parents[0]
data_path = home/'data'/'raw'
data_train = data_path/'jpeg'/'train'
data_test = data_path/'jpeg'/'test'

print(data_train)
print(data_test)
```

```python
# list of all pictures in train/test directory
ftr = []
for (dirpath, dirnames, filenames) in walk(data_train):
    ftr.extend(filenames)
    break
first_ftr = ftr[0]
print(first_ftr)

fte = []
for (dirpath, dirnames, filenames) in walk(data_test):
    fte.extend(filenames)
    break
first_fte = fte[0]

print(first_fte)
```

```python
# list sizes in train directory
size_tr = list(map(lambda x: im.open(data_train/x).size,ftr))
size_tr_uniq = set(size_tr) #convert to set to delete duplicates

name_size_tr = list(map(lambda x: (x,im.open(data_train/x).size),ftr))
```

```python
results = [t[0] for t in name_size_tr if t[1] == (6000,4000)]
print(results)
```

```python
train_df = pd.read_csv(data_path/'train.csv')
print(train_df.head())
```

```python
results2 = [s.replace('.jpg', '') for s in results] # remove '.jpg' 
df_6000_4000 = train_df[train_df.image_name.isin(results2)] # find images with (6000,4000)

print(df_6000_4000.groupby('target').target.count()) # number of target=1/malignant diasnosis in this subset
```

```python
# list sizes in test directory
size_te = list(map(lambda x: im.open(data_test/x).size,fte))
size_te_uniq = set(size_te) #convert to set to delete duplicates
print(size_te_uniq)
```

```python
# scatter plot of file sizes
x_train,y_train = zip(*size_tr)
plt.scatter(x_train,y_train)
```

```python
# count of unique in train. Looks like top 10 constitute >99%.
count_uniq_train = Counter(elem for elem in size_tr)
print(count_uniq_train)
```

```python
x_test,y_test = zip(*size_te)
plt.scatter(x_test,y_test)
```

```python
# count of unique in test. Looks like top 10 constitute >99%.
count_uniq_test = Counter(elem for elem in size_te)
print(count_uniq_test)
```
